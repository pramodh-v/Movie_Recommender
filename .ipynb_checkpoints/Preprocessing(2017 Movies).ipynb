```python
g_null = dataset['genres']!='[]'
genres = dataset['genres'][g_null]
```

```python
genres.head()
```

```python
import ast
```

```python
def gen_to_list(row):
    row = ast.literal_eval(row)
    return [g['name'] for g in row]
```

```python
genres_list = genres.apply(gen_to_list)
genres_list.head()
```

```python
def list_to_string(row):
    ret_str = ''
    for stri in row:
        if(stri=='Science Fiction'):
            stri = 'Sci-Fi'
        ret_str+=stri+' '
    return ret_str
```

```python
genres_list_str = genres_list.apply(list_to_string)
genres_list_str
```

```python
dataset['genres'] = genres_list_str
```

```python
dataset
```

```python
credits = pd.read_csv('credits.csv')
```

```python
credits.head(10)
```

```python
credits.shape
```

```python
dataset = dataset[['id','title','genres','year']]
```

```python
dataset.head()
```

```python
dataset['id'] = dataset['id'].astype(int)
```

```python
dataset = pd.merge(dataset,credits,on='id')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset['cast'] = dataset['cast'].map(lambda x: ast.literal_eval(x))
dataset['crew'] = dataset['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
dataset['actor_1_name'] = dataset['cast'].apply(get_actor1)
```

```python
dataset['actor_2_name'] = dataset['cast'].apply(get_actor2)
```

```python
dataset['actor_3_name'] = dataset['cast'].apply(get_actor3)
```

```python
dataset
```

```python
def getdir(row):
    dt = []
    st = " "
    for i in row:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
dataset['director'] = dataset['crew'].apply(getdir)
```

```python
dataset = dataset[['director','actor_1_name','actor_2_name','actor_3_name','genres','title']]
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.dropna()
```

```python
dataset = dataset.rename(columns={'title':'movie_title'})
```

```python
dataset.to_csv('movies_2017.csv',index=False)
```
