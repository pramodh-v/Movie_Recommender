```python
import pandas as pd
import numpy as np
```

```python
dataset= pd.read_csv('IMDb Movies.csv')
dataset.head()
```

```python
dataset = dataset.loc[(dataset.year==2018)|(dataset.year==2019)|(dataset.year==2020)]
```

```python
dataset
```

```python
dataset = dataset[['director','actors','genre','title']]
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.dropna()
```

```python
actor_1_name=[]
actor_2_name=[]
actor_3_name=[]
for actor in dataset['actors']:
    if(len(actor.split(','))<1):
        actor_1_name.append('Unknown')
        actor_2_name.append('Unknown')
        actor_3_name.append('Unknown')
    elif(len(actor.split(','))<2):
        actor_1_name.append(actor.split(',')[0])
        actor_2_name.append('Unknown')
        actor_3_name.append('Unknown')
    elif(len(actor.split(','))<3):
        actor_1_name.append(actor.split(',')[0])
        actor_2_name.append(actor.split(',')[1])
        actor_3_name.append('Unknown')
    else:
        actor_1_name.append(actor.split(',')[0])
        actor_2_name.append(actor.split(',')[1])
        actor_3_name.append(actor.split(',')[2])
```

```python
dataset['actor_1_name'] = actor_1_name
dataset['actor_2_name'] = actor_2_name
dataset['actor_3_name'] = actor_3_name
```

```python
dataset
```

```python
dataset['genre'] = dataset['genre'].str.replace(',',' ')
```

```python
director_ = []
for director in dataset['director']:
    director_.append(director.split(',')[0])
```

```python
dataset['director'] = director_
```

```python
dataset
```

```python
dataset = dataset[['director','actor_1_name','actor_2_name','actor_3_name','genre','title']]
```

```python
dataset = dataset.rename(columns={'genre':'genres'})
dataset = dataset.rename(columns={'title':'movie_title'})
```

```python
dataset.head(20)
```

```python
dataset.to_csv('movies_2018-2020.csv')
```
