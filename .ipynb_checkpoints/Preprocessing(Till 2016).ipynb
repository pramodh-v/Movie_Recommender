```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('movie_metadata.csv')
```

```python
dataset.head(15)
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
dataset['title_year'].value_counts().sort_index().plot(kind='barh',figsize=(20,20))
```

```python
for i in dataset.columns:
    print(i,dataset[i].isnull().sum())
```

```python
dataset = dataset[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
dataset
```

```python
for row in ['actor_1_name','actor_2_name','actor_3_name','director_name']:
    dataset[row] = dataset[row].replace(np.nan,'Unknown')
```

```python
dataset['genres'] = dataset['genres'].str.replace('|',' ')
```

```python
dataset['movie_title'][0]
```

```python
dataset['movie_title'] = dataset['movie_title'].apply(lambda x:x[:-1])
```

```python
dataset
```

```python
dataset = dataset.rename(columns={'director_name':'director'})
```

```python
dataset.to_csv('movie_2016.csv')
```
